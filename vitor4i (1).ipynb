```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.style.use('ggplot')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

file_path = "vendas_distribuidoras_anp.xlsx"
diesel = pd.read_excel(file_path, sheet_name="diesel")
gasolina = pd.read_excel(file_path, sheet_name="gasolina")
etanol = pd.read_excel(file_path, sheet_name="etanol")

print("Diesel - Primeiras linhas:")
display(diesel.head())
print("\nGasolina - Primeiras linhas:")
display(gasolina.head())
print("\nEtanol - Primeiras linhas:")
display(etanol.head())
```

```python
def melt_df(df, combustivel):
    
    df_melted = df.melt(
        id_vars=["regiao", "meses"], 
        var_name="ano", 
        value_name="vendas"
    )
    df_melted["combustivel"] = combustivel
    return df_melted


df_full = pd.concat([
    melt_df(diesel, "diesel"),
    melt_df(gasolina, "gasolina"),
    melt_df(etanol, "etanol")
])


df_full["ano"] = df_full["ano"].astype(int)
df_full.dropna(subset=["vendas"], inplace=True)


print("\nDados consolidados - Primeiras linhas:")
display(df_full.head())
```

```python
sazonalidade = df_full.groupby(["meses", "combustivel"])["vendas"].mean().unstack()


sazonalidade.plot(marker="o", figsize=(14, 6))
plt.title("Sazonalidade das Vendas de CombustÃ­veis (MÃ©dia 2000-2021)")
plt.ylabel("Vendas (milhÃµes de litros)")
plt.xticks(range(1, 13), ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"])
plt.grid(True)
plt.show()
```

```python
diesel_sp_ma = df_full[
    (df_full["combustivel"] == "diesel") & 
    (df_full["regiao"].isin(["sp", "ma"]))
].groupby(["regiao", "meses"])["vendas"].mean().unstack()


diesel_sp_ma.T.plot(marker="o", figsize=(14, 6))
plt.title("Sazonalidade de Vendas de Diesel: SP vs MA (MÃ©dia 2000-2021)")
plt.ylabel("Vendas (milhÃµes de litros)")
plt.xticks(range(1, 13), ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"])
plt.grid(True)
plt.show()
```

```python

distrib_2021 = df_full[df_full["ano"] == 2021].groupby(["regiao", "combustivel"])["vendas"].sum().unstack()


distrib_2021.plot(kind="bar", stacked=True, figsize=(14, 6))
plt.title("DistribuiÃ§Ã£o do Consumo por Estado e CombustÃ­vel (2021)")
plt.ylabel("Vendas (milhÃµes de litros)")
plt.xticks(rotation=45)
plt.show()
```

```python
mercado_5anos = df_full[df_full["ano"].between(2017, 2021)].groupby(["ano", "combustivel"])["vendas"].sum().unstack()


mercado_5anos.plot(marker="o", figsize=(14, 6))
plt.title("EvoluÃ§Ã£o do Mercado de CombustÃ­veis (2017-2021)")
plt.ylabel("Vendas (bilhÃµes de litros)")
plt.grid(True)
plt.show()
```
